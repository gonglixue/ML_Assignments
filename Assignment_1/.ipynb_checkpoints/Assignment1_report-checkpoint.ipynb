## 3. Gaussian Discriminant Analysis and MLE
### (a)
$ p(y=1 \big| x; \phi, \mu_{0}, \mu_{1}, \Sigma_{0}, \Sigma_{1}) =  p(x \big|y=1) * \phi $
$ = \frac{1}{4\pi} e^{-\frac{1}{2}(x-\mu_{1})^{T}  (x-\mu_{1})}$

$ p(y=0 \big| x; \phi, \mu_{0}, \mu_{1}, \Sigma_{0}, \Sigma_{1}) = p(x \big| y=0) * (1-\phi) $
$ = \frac{1}{4\pi} e^{-\frac{1}{2} (x-\mu_{0})^{T} (x-\mu_{0})} $  

So the decision boundary is:
$ (x_{1} - 1)^{2} + (x_{2}-1)^{2} = (x_{1}-0)^{2} + (x_{2} - 0)^{2} $

namely, $ x_{1} + x_{2} - 1 = 0 $

In my opinion, for a spam filter, precision is more important. It is not a matter if a user receive only a few of spam emails. While users don't want their important emails are lost.
